```python
import pandas as pd
```

```python
df = pd.read_csv('Dataset of Diabetes .csv')
df
```

```python
df = df.drop(columns=['ID'])
df
```

```python
df['Gender'] = df['Gender'].map({'M': 1, 'F': 0}).astype(bool)
df['CLASS'] = df['CLASS'].map({'Y': 0, 'Y ': 0,
                               'N': 1, 'N ': 1,
                               'P':2})
df
```

```python
#Standard scalar
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
dados_normalizados = pd.DataFrame(scaler.fit_transform(df.drop
(columns=['CLASS'])),
                                  columns=df.drop(columns=['CLASS']).columns)
dados_normalizados['CLASS'] = df['CLASS'].values
dados_normalizados
```

```python
dados_normalizados.describe()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca.fit(dados_normalizados.drop(columns=['CLASS']))
dados = pd.DataFrame(pca.transform(df.drop(columns=['CLASS'])),
                     columns=['PC1', 'PC2'])
dados['Classe'] = df['CLASS'].values
print(pca.explained_variance_ratio_)
```

```python
dados
```

```python
#plot 2d com a classe

import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,6))
sns.scatterplot(data=dados, x='PC1', y='PC2', hue='Classe', palette='viridis')
plt.title('PCA - Diabetes Dataset')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend(title='Classe')
plt.show()
```
